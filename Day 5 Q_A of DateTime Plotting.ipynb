```
import pandas as pd

df = pd.DataFrame({'values': [4,2,7,4,9,7]})

df.head()
```

```
df['date'] = ['FEB-6-2020', 'FEB-7-2020', 'FEB-8-2020', 'FEB-9-2020', 'FEB-10-2020', 'FEB-11-2020']

df.head(6)
```

### Solution 1, live with ugly tick labels

```
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.plot(df['values'])
ax.set_xticklabels(df['date'])

plt.show()
```

## Solution 2, convert tick labels to dates

```
df['date'] = pd.to_datetime(df['date'])

df.head()
```

## still looks ugly

```
fig, ax = plt.subplots()

ax.plot(df['values'])
ax.set_xticklabels(df['date'])

plt.show()
```

```
df['month'] = (df['date'].dt.month) 
df['day'] = (df['date'].dt.day)

df.head()
```

```
df.dtypes
```

```
df['pretty_date'] = df['month'].astype(str) + '/' + df['day'].astype(str)
  
df.head()
```

```
fig, ax = plt.subplots()

ax.plot(df['values'])
ax.set_xticklabels(df['pretty_date'])

plt.show()
```

```
def pretty_month(cell_contents):
  if cell_contents == 2:
    return "Feb"

df['month_string'] = df['month'].apply(pretty_month)

df.head()
```

```
df['month_string_day'] = df['month_string'] + " " + df['day'].astype(str)


df.head()
```

```
fig, ax = plt.subplots()

ax.plot(df['values'])
ax.set_xticklabels(df['month_string_day'])

plt.show()
```
